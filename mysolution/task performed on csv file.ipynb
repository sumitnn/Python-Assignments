```python
# import libraries for this task
import pandas as pd
import numpy as np
import re
import warnings
warnings.simplefilter("ignore")
```

```python
# load movies.csv and comment file and not load unnamed column
# df consider movies collection and df1 consider comment collection
df=pd.read_csv('movies.csv').iloc[:, 1:]
df1=pd.read_csv('comment.csv').iloc[:, 1:]
#show all columns and records
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
# top 5 rows
df.head(2)
```

```python
# dataframe movies_with_comments 
movies_with_comments=df[~(df['num_mflix_comments'].isna())]
movies_with_comments.sample(3)
```

```python
# movies with no comments
movies_with_no_comments=df[df['num_mflix_comments'].isna()]
movies_with_no_comments.head(3)
```

```python
# create movies_with_no_comments.csv and save into a movies folder
movies_with_no_comments.to_csv('movies/movies_with_no_comments.csv',index = False)
```

```python
# create two new columns low_runtime and high_runtime 
# create function gtfifty
def gtfifty(x):
    if x>50:
        return 'Yes'
    return 'No'
# create function ltfifty   
def ltfifty(x):
    if x>50:
        return 'No'
    return 'Yes'
movies_with_comments['low_runtime']=movies_with_comments['runtime'].apply(gtfifty)
movies_with_comments['high_runtime']=movies_with_comments['runtime'].apply(ltfifty)
# top 5 rows
movies_with_comments.head()
```

```python
#  save file in movies folder
movies_with_comments.to_csv('movies/movies_with_comments.csv',index =False)
```

```python
# get rating value from obj,dictionary and create new Rating column and nan values fill with 0.0
def fun(x):
    try:
        x=eval(x)
        x=x.get('rating')
        x=float(x)
    except :
        x=0.0

    return x
    
df['Rating']=df['imdb'].apply(fun)
```

```python
# get win award value from obj,dictionary and create new TotalAwards column and nan values fill with 0
def fun(x):
    try:
        x=eval(x)
        x=x.get('wins')
        x=int(x)
    except :
        x=0

    return x
    
df['TotalAwards']=df['awards'].apply(fun)
```

```python
# create function to convert str to int and handle invalid literal also
def contoint(x):
    try:
        return int(x)
    except:
        x=re.findall(r'\d+',x)[0]
        return int(x)
    
df['year']=df['year'].apply(lambda x : contoint(x))
```

```python
# Using movies collection fetch all movies having imdb rating > 8 and released on or after
# year 2000 with more than 3 awards in ascending order of released.Save the dataframe
# in movies directory with filename movies_rating_8_released_aft_2000.csv
filterdata=df[(df['Rating']>8) & (df['year'] >=2000) & (df['TotalAwards']>3)]
filterdata.head(2)
```

```python
# change datatype of  realeased column into datetime 
filterdata['released']=  pd.to_datetime(filterdata['released'])
filterdata=filterdata.sort_values('released',ascending=True)
```

```python
# create csv file and drop that column which i made for filtering
filterdata.drop(['TotalAwards','Rating'],axis=1,inplace=True)
filterdata.to_csv('movies/movies_rating_8_released_aft_2000.csv',index=False)
```

```python
# Using movies collection, create a new dataframe released_outside_usa where countries
# value should not be equal to USA . Save the dataframe in movies directory with filename
# released_outside_usa.csv 
# get country value from obj,list and create column Countries fill usa with nan in that column
def fun(x):
    try: 
        
        if 'USA' in x:
            x=np.nan
        else:
            x=x
            
    except :
        x=np.nan
    

    return x
    
df['Countries']=df['countries'].apply(fun)



# released_outside_usa=
```

```python
released_outside_usa=df[~(df['Countries'].isna())]
```

```python
# drop that column which i made 
released_outside_usa.drop(['Countries','TotalAwards','Rating'],axis=1,inplace=True)
```

```python
# save file
released_outside_usa.to_csv('movies/released_outside_usa.csv',index=False)
```
